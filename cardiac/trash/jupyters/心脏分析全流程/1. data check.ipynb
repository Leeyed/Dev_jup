```python
import os
import SimpleITK as sitk
import matplotlib.pyplot as plt
```

```python
demp_dir = r'./demo_5samples'

file_list = list(filter(lambda x: x.endswith('.nii.gz'), os.listdir(demp_dir)))
print(file_list)
```

```python
fig = plt.figure(figsize=(12,8))

for ind,filename in enumerate(file_list):
    arr_img = sitk.GetArrayFromImage(sitk.ReadImage(os.path.join(demp_dir, filename)))
    # print(arr_img.shape)
    ax = plt.subplot(2,3, ind+1)
    ax.set_axis_off()
    ax.set_title(filename)
    plt.imshow(arr_img[30,:,:], vmin=0, vmax=20)
    plt.colorbar()
```
